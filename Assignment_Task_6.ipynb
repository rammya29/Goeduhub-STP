 3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

     `days=[1,2,3,4,5,6,7,8]`

 `max_speed=[80,91,92,88,77,79,76,75]`
 
 `min_speed=[42,43,40,42,33,36,34,35]`

 `avg_speed=[46,58,57,56,40,42,41,36]`

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]

plt.plot(days, max_speed, ":", label = "Maximum_Speed")
plt.plot(days, min_speed, ":", label = "Minimum_Speed")
plt.plot(days, avg_speed, ":", label = "Average_Speed")

plt.grid()
plt.legend()

plt.title("Car Speed Measurement")
plt.show()
```

4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

5. Plot Simple bar chart showing popularity of Programming Languages.

 `Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']`

 `Popularity = [56, 39, 34, 34, 29]`

 `Security = [44 ,36 ,55, 50, 42]`

 Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
import matplotlib.pyplot as plt

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]

plt.bar(Languages, Popularity,  label = "Popularity")
plt.bar(Languages, Security, width = 0.4, label = "Security")
plt.title("Language")
plt.legend(loc='upper right', frameon=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]

plt.barh(Languages, Popularity,  label = "Popularity")
plt.barh(Languages, Security, height = 0.4, label = "Security")
plt.title("Language")
plt.legend(loc='upper right', frameon=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/titanic.csv')

data_male = data[data['Sex'] == 'male']
data_male.Sex.value_counts()

data_female = data[data['Sex'] == 'female']
data_female.Sex.value_counts()

plt.figure(figsize = (10,5)) 
plt.title("Gender")
plt.xlabel("Fare")
plt.ylabel("Age")

plt.scatter(data_male['Fare'], data_male['Age'], label = "Male", color = 'r', marker = '*' )
plt.scatter(data_female['Fare'], data_female['Age'], label = "Female", color = 'm', marker = '+' )
 

plt.legend()
```
